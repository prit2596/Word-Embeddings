```python
import nltk
import tensorflow as tf 
import numpy as np

nltk.download("punkt")

sentence = "He is the king. The king is royal. She is the royal queen."
sentence = sentence.lower()
print(sentence)
to = nltk.word_tokenize(sentence)
```

```python
ts = set(to)
tm = list(ts)
t = []
for i in range(0, len(tm)):
    if tm[i] != ".":
        t.append(tm[i])
sent = sentence.split('.')

markers= []
for js in sent:
        markers.append(js.split())
markers.remove([])
print(markers)
data= []
for sent in markers:
    for word_index, word in enumerate(sent):
        for n_word in sent[max(word_index-2,0):min(word_index+2,len(sent))+1]:
            if n_word != word:
                data.append([word, n_word])
                
w2i = {}
i2w = {}
print(data)
for i in range(0, len(t)):
    w2i[t[i]] = i
    i2w[i] = t[i]

vocab_size = len(t)

def to_one_hot(data_point_index, vocab_size):
    temp = np.zeros(vocab_size)
    temp[data_point_index] = 1
    return temp

x_train = []
y_train = []

for d in data:
    x_train.append(to_one_hot(w2i[d[0]], vocab_size))
    y_train.append(to_one_hot(w2i[d[1]], vocab_size))
    
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)
```

```python
print(x_train)
```

```python
print(x_train.shape, y_train.shape)
```

```python
x = tf.placeholder(tf.float32, shape=(None, vocab_size))
y_label = tf.placeholder(tf.float32, shape=(None, vocab_size))
print(x, y_label)
```

```python
EMBEDDING_DIM = 7

W1 = tf.Variable(tf.random_normal([vocab_size, EMBEDDING_DIM]))
print(W1)
```

```python
# Bias
b1 = tf.Variable(tf.random_normal([EMBEDDING_DIM]))
print(b1)
```

```python
hidden_representation = tf.add(tf.matmul(x, W1), b1)
print(hidden_representation)
```

```python
W2 = tf.Variable(tf.random_normal([EMBEDDING_DIM, vocab_size]))
print(W2)
```

```python
b2 = tf.Variable(tf.random_normal([vocab_size]))
print(W2)
```

```python
prediction = tf.nn.softmax(tf.add( tf.matmul(hidden_representation, W2), b2))
print(prediction)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init) 

# loss fn
cross_entropy_loss = tf.reduce_mean(-tf.reduce_sum(y_label * tf.log(prediction), reduction_indices=[1]))

train_step = tf.train.GradientDescentOptimizer(0.9).minimize(cross_entropy_loss)
epoch = 10000

# train
for _ in range(epoch):
    sess.run(train_step, feed_dict={x: x_train, y_label: y_train})
    print('loss is : ', sess.run(cross_entropy_loss, feed_dict={x: x_train, y_label: y_train}))
```

```python
print(sess.run(W1))
print('----------')
print(sess.run(b1))
print('----------')
```

```python
vectors = sess.run(W1 + b1)
```

```python
print(vectors)
```

```python
def euclidean_dist(vec1, vec2):
    return np.sqrt(np.sum((vec1-vec2)**2))

def find_closest(word_index, vectors):
    min_dist = 9999
    min_index = -1
    
    query_vector = vectors[word_index]
    
    for index, vector in enumerate(vectors):
        if euclidean_dist(vector, query_vector) < min_dist and not np.array_equal(vector, query_vector):
            min_dist = euclidean_dist(vector, query_vector)
            min_index = index
    return min_index
```

```python
print(i2w[find_closest(w2i['king'], vectors)])
```
