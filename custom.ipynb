```python
import nltk
nltk.download("punkt")
sentence = "He is a royal . The king is a man . The king has a queen."
to = nltk.word_tokenize(sentence)
```

```python
ts = set(to)
tm = list(ts)
t = []
for i in range(0, len(tm)):
    if tm[i] != ".":
        t.append(tm[i])
sent = sentence.split('.')

markers= []
for js in sent:
        markers.append(js.split())
markers.remove([])
print(markers)
k = []
for sent in markers:
    for word_index, word in enumerate(sent):
        for n_word in sent[max(word_index-2,0):min(word_index+2,len(sent)+1)]:
            if n_word != word:
                k.append([word, n_word])

print(k)
w2i = {}
i2w = {}

for i in range(0, len(t)):
    w2i[t[i]] = i
    i2w[i] = t[i]

print(w2i)

hot = []
for i in range(0, len(t)):
    new_arr = [0]*len(t)
    x = w2i[t[i]]
    new_arr[x] = 1
    hot.append(new_arr)
print(hot)
```
